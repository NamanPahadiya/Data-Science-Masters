```python
#Ans3:
""""The ArithmeticError class is the base class for all errors related to arithmetic operations. 
It serves as a parent class for more specific arithmetic-related error classes."""

#ZeroDivisionError
"""The ZeroDivisionError is raised when a division performed with zero as the divisor"""
try:
    a=10
    a/0
except ZeroDivisionError as e:
    print("Error::",e)
```

```python
#OverflowError
"""The OverflowError is raised when an arithmetic operation results in a value that is too 
large to be represented within the numeric type being used"""
import math

try:
    value=math.exp(1000)

#The result exceeds the maximum representable value for a floating-point number, leading to an OverflowError
except OverflowError as e:
    print("Error:",e)
```

```python
#Ans4:
""""The LookupError class in Python is the base class for errors related to accessing or looking 
up elements in a sequence or mapping (dictionary-like) structure"""

#KeyError
"""The KeyError is raised when you try to access a key in a dictionary or a similar mapping structure that does not exist."""

dict_1={"a":1,"b":2}
try:
    value=dict_1["alphabets"]
except KeyError as e:
    print("Error:",e)
```

```python
#IndexError
"""The IndexError is raised when you try to access an element in a sequence (such as a list or tuple) using an index that is out of range."""
l=[1,2,3,4,5,6,]
try:
    l[10]
except IndexError as e:
    print("Error",e)
```
